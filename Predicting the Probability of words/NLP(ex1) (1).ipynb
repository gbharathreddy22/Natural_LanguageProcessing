```python
f=open('text.txt','r')
s=f.read()
import nltk
from nltk.tokenize import word_tokenize
d=word_tokenize(s)
for i in d:
    if i== '.':
        d.remove('.')
#print(len(d))
s=[]
for i in d:
    if i not in s:
        s.append(i)
#print(s)
#print(len(s))
l=[]
for i in range(len(d)-1):
    a=d[i]
    b=d[i+1]
    c=0
    for j in range(len(d)-1):
        if a==d[j] and b==d[j+1]:
            c=c+1
    p=[a,b]
    q=[p,c]
    l.append(q)
#print(len(l))
x=[]
for i in l:
    if i not in x:
        x.append(i)
#print(x)
#print(len(x))
y=[]
z=[]
for i in range(len(x)):
    b=x[i]
    v=b[1]/len(d)
    o=[b[0],v]
    y.append(o)
    z.append(v)
    print("Probability of",b[0],"is : ",v)
mx=max(z)
print("MAX PROBABLE PAIR:")
for i in y:
    if mx==i[1]:
        print("Probability of",i[0],"is : ",i[1])
mn=min(z)
print("MIN PROBABLE PAIR:")
for i in y:
    if mn==i[1]:
        print("Probability of",i[0],"is : ",i[1])
```
